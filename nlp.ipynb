```python
from sklearn.feature_extraction.text import CountVectorizer

texto = ['Assisti um filme ótimo', 'Assisti um filme ruim']

vectorizer = CountVectorizer(lowercase=False)
bag_of_words = vectorizer.fit_transform(texto)
bag_of_words
```

```python
vocabulario = vectorizer.get_feature_names() #conjunto de palavras a ser analizadas no frame
vocabulario
```

```python
matriz_esparsa = pd.SparseDataFrame(bag_of_words, columns=vocabulario)
matriz_esparsa
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

vectorizer = CountVectorizer(lowercase=False, max_features=50) #max_features setta a quantia de palavras para o vocabulario de cada resenha, pegando assim as (nesse caso) 50 mais recorrentes
bag_of_words = vectorizer.fit_transform(resenha.text_pt)

treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words,
                                                              resenha.classificacao,
                                                              random_state=48)
model = LogisticRegression()
model.fit(treino, classe_treino)
acuracia = model.score(teste, classe_teste)
print(f'Acuracia final: {round(acuracia * 100, 2)}%')
```

```python
def classifica_resenhas(data_frame, coluna_texto, coluna_classificacao):
    vectorizer = CountVectorizer(lowercase=False, max_features=50)
    bag_of_words = vectorizer.fit_transform(data_frame[coluna_texto])

    treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words,
                                                                  data_frame[coluna_classificacao],
                                                                  random_state=48)
    model = LogisticRegression(solver='lbfgs') #solver pro warning sumir
    model.fit(treino, classe_treino)
    acuracia = model.score(teste, classe_teste)
    print(f'Acuracia final: {round(acuracia * 100, 2)}%')
```

```python
classifica_resenhas(resenha, 'text_pt', 'classificacao')
```

```python
def nuvem_negativa(data_frame, coluna_texto):
    texto_negativo = data_frame.query('classificacao == 0')
    todas_as_palavras = ' '.join([texto for texto in texto_negativo[coluna_texto]])
    nuvem_de_palavras = WordCloud(width=800, height=500, 
                                  max_font_size=110, collocations=False).generate(todas_as_palavras) #collocations retira bigramas

    plt.figure(figsize=(12, 8))
    plt.imshow(nuvem_de_palavras, interpolation='bilinear') #interpolation aumenta a nitidez
    plt.axis('off')
    plt.show()
```

```python
def plot_maiores_ocorrencias(data_frame, coluna_texto, quantidade):
    todas_as_palavras = ' '.join([texto for texto in data_frame[coluna_texto]])
    
    tokenizer = tokenize.WhitespaceTokenizer()
    todas_as_palavras_tokenized = tokenizer.tokenize(todas_as_palavras)
    frequencia = nltk.FreqDist(todas_as_palavras_tokenized)
    
    tokenized_data_frame = pd.DataFrame(
        {'palavra': list(frequencia.keys()), 
         'frequencia': list(frequencia.values())
        }
    )
    maior_ocorrencia = tokenized_data_frame.nlargest(columns='frequencia', n=quantidade)
    plt.figure(figsize=(12, 8))
    sns.barplot(data=maior_ocorrencia, x='palavra', y='frequencia')
    plt.show()
```

```python
plot_maiores_ocorrencias(resenha, 'text_pt', 15)
```

```python
palavras_irrelevantes = nltk.corpus.stopwords.words('portuguese')

frase_processada = list()

for opiniao in resenha.text_pt:
    nova_frase = list()
    whitespace_tokenizer = tokenize.WhitespaceTokenizer()
    palavras_tokenized = whitespace_tokenizer.tokenize(opiniao)
    for palavra in palavras_tokenized:
        if palavra not in palavras_irrelevantes:
            nova_frase.append(palavra)
    frase_processada.append(' '.join(nova_frase))
    
resenha['tratamento_1'] = frase_processada
```

```python
from string import punctuation

pontuacao = [ponto for ponto in punctuation]

palavras_irrelevantes += pontuacao

opnioes_processadas = list()
for opniao in resenha['tratamento_1']:
    opniao_sem_pontuacao = list()
    opniao_processada = tokenize.WordPunctTokenizer().tokenize(opniao)
    for letra in opniao_processada:
        if letra not in palavras_irrelevantes:
            opniao_sem_pontuacao.append(letra)
    opnioes_processadas.append(' '.join(opniao_sem_pontuacao))

resenha['tratamento_2'] = opnioes_processadas
```

```python
import unidecode

palavras_irrelevantes = [unidecode.unidecode(opniao) for opniao in palavras_irrelevantes]
resenha['tratamento_3'] = [unidecode.unidecode(opniao) for opniao in resenha['tratamento_2']]

opnioes_processadas = list()
for opniao in resenha['tratamento_3']:
    opniao_sem_pontuacao = list()
    opniao_processada = tokenize.WordPunctTokenizer().tokenize(opniao)
    for palavra in opniao_processada:
        if palavra not in palavras_irrelevantes:
            opniao_sem_pontuacao.append(palavra)
    opnioes_processadas.append(' '.join(opniao_sem_pontuacao))

resenha['tratamento_3'] = opnioes_processadas
```

```python
palavras_irrelevantes = [opniao.lower() for opniao in palavras_irrelevantes]
resenha['tratamento_4'] = [opniao.lower() for opniao in resenha['tratamento_3']]

opnioes_processadas = list()
for opniao in resenha['tratamento_4']:
    opniao_sem_pontuacao = list()
    opniao_processada = tokenize.WordPunctTokenizer().tokenize(opniao)
    for palavra in opniao_processada:
        if palavra not in palavras_irrelevantes:
            opniao_sem_pontuacao.append(palavra)
    opnioes_processadas.append(' '.join(opniao_sem_pontuacao))

resenha['tratamento_4'] = opnioes_processadas
```

```python
stemmer = nltk.RSLPStemmer() #retira flexoes encontrando a raiz em comum

opnioes_processadas = list()
for opniao in resenha['tratamento_4']:
    opniao_sem_pontuacao = list()
    opniao_processada = tokenize.WordPunctTokenizer().tokenize(opniao)
    for palavra in opniao_processada:
        if palavra not in palavras_irrelevantes:
            opniao_sem_pontuacao.append(stemmer.stem(palavra))
    opnioes_processadas.append(' '.join(opniao_sem_pontuacao))

resenha['tratamento_5'] = opnioes_processadas
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#tfidf modela 'pesos' para as palavras de acordo com sua ocorrencia, tipo o que o wordcloud faz

def classifica_resenhas_tfidf(data_frame, coluna_texto, coluna_classificacao):
    tfidf = TfidfVectorizer(lowercase=False, max_features=50)
    bag_of_words = tfidf.fit_transform(data_frame[coluna_texto])

    treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words,
                                                                  data_frame[coluna_classificacao],
                                                                  random_state=42)
    model = LogisticRegression(solver='lbfgs') #solver pro warning sumir
    model.fit(treino, classe_treino)
    acuracia = model.score(teste, classe_teste)
    print(f'Acuracia final: {round(acuracia * 100, 2)}%')
```

```python
classifica_resenhas_tfidf(resenha, 'tratamento_5', 'classificacao')
```

```python
def classifica_resenhas_tfidf_ngrams_semlimite(data_frame, coluna_texto, coluna_classificacao):
    tfidf = TfidfVectorizer(lowercase=False, ngram_range=(1, 2))
    bag_of_words = tfidf.fit_transform(data_frame[coluna_texto])

    treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words,
                                                                  data_frame[coluna_classificacao],
                                                                  random_state=42)
    model = LogisticRegression(solver='lbfgs') #solver pro warning sumir
    model.fit(treino, classe_treino)
    acuracia = model.score(teste, classe_teste)
    print(f'Acuracia final: {round(acuracia * 100, 2)}%')
```

```python
classifica_resenhas_tfidf_ngrams_semlimite(resenha, 'tratamento_5', 'classificacao')
```
